### Analysis of the dataset Housing
The following file will present an analysis of the dataset and present the presentation by text, data manipulation and by visualization.

```python
import pandas as pd

houses = pd.read_csv("housing.csv")

print("\nBy printing the first 8 rows of data we can see that the housing data are sorted by ....\nno particular order.\n\n")
print(houses.head(8))
print("\n\nWe can print the last 8 rows of data to see that the last houses are ....\nin no special order either.\n\n")
print(houses.tail(8))
print("\n\nWe can display the info of the housing and see how many non-null rows each column have!\n\n")
houses.info()
print("\n\nLets drop the rows which are missing column values!\nNow we can see the difference below.\nWe went from 20640 entries to 20433.\n\n")
houses.dropna(inplace=True)  
houses.info()
```

```python
import matplotlib.pyplot as plt

print("We have the following properties to sort by:\n- longitude\n- latitude\n- housing_median_age\n- total_rooms\n- total_bedrooms\n- population\n- households\n- median_income\n- median_house_value\n- ocean_proximity")
print("\nLets create some visual data.")
print("By sort the housing data by latitude and longitude aswell as population, we can clearly see the location of populations of people.")

plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    houses["longitude"],
    houses["latitude"],
    c=houses["population"],
    cmap="hot",  
    alpha=0.5,
    s=5
)
plt.colorbar(scatter, label="Population")
plt.title("Mapping the populations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```

```python
sorted_rooms = houses.sort_values(by="total_rooms", ascending=False)
sorted_bedrooms = houses.sort_values(by="total_bedrooms", ascending=False)

max_rooms = houses["total_rooms"].max()
max_bed_rooms = houses["total_bedrooms"].max()
print("It could be interesting to also see the max of the amount of bedrooms and rooms in the dataset.")
print("The most amount of bedrooms: ", int(max_bed_rooms))
print("The most amount of rooms: ", int(max_rooms))
print("Must be the billionaires and hotels..")
 
print("\nIf we use the properties ocean_proximity, households, population we can clearly see how popular it is to live by the sea.")
sea_h = houses.groupby("ocean_proximity")["population"].sum()

sea_h_sorted = sea_h.sort_values(ascending=False)

plt.figure(figsize=(10, 5))
bars = plt.bar(
    sea_h_sorted.index,
    sea_h_sorted.values,
    color=["blue", "lightblue", "turquoise", "lightblue", "blue"][:len(sea_h_sorted)]
)
plt.title("Population ordered by ocean proximity")
plt.xlabel("Ocean proximity")
plt.ylabel("Population")
plt.tight_layout()
plt.show()
```

```python
print("We can also compare median_income with median_house_value to see the differences of income/house-evaluation.\nVery interesting.")  
plt.figure(figsize=(10, 5))
plt.scatter(
    houses["median_income"],
    houses["median_house_value"],
    alpha=0.2,
    color="blue"
)
plt.title("Presentation of median income and median house value")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("By displaying households and sorting by total_bedrooms and median_income we can see how the income correlates with amount of bedrooms")
plt.figure(figsize=(10, 5))
plt.scatter(
    houses["total_bedrooms"],
    houses["median_income"],
    alpha=0.3,
    color="blue",
    label="Bedrooms"
)
plt.title("Amount of bedrooms ordered by median income")
plt.xlabel("Bedrooms")
plt.ylabel("Median Income")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
